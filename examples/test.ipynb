```python
import pandas as pd

# Replace 'your_path_here' with the path to your CSV file
file_path = "../data/Traffic_Crashes_-_People.csv"

# Read the CSV file
df1 = pd.read_csv(file_path)

# Display information about the DataFrame
df1.info()
```

```python

# Convert date column
df1['CRASH_DATE'] = pd.to_datetime(df1['CRASH_DATE'], errors='coerce')

# Convert integer columns with missing values to nullable integer type
df1['VEHICLE_ID'] = df1['VEHICLE_ID'].astype('Int64')
df1['SEAT_NO'] = df1['SEAT_NO'].astype('Int64')
df1['AGE'] = df1['AGE'].astype('Int64')

df1.info()
```

```python
import pandas as pd

# Replace 'your_path_here' with the path to your CSV file
file_path = "../data/Traffic_Crashes_-_Vehicles.csv"

# Read the CSV file
df2 = pd.read_csv(file_path)

# Display information about the DataFrame
df2.info()
```

```python
df2['CRASH_DATE'] = pd.to_datetime(df2['CRASH_DATE'], errors='coerce')
df2['NUM_PASSENGERS'] = df2['NUM_PASSENGERS'].astype('Int64')
df2['UNIT_TYPE'] = df2['UNIT_TYPE'].astype('category')
```

```python
df2.info()
```

```python
import matplotlib.pyplot as plt
# Calculate total and null values for each column
total_values = df2.shape[0]
null_values = df2.isnull().sum()
non_null_values = total_values - null_values
# Print the total and null values for each column
column_info = pd.DataFrame({
    # 'Total Values': total_values,
    'Non-Null Values': non_null_values,
    'Null Values': null_values
})

# Set display options to show all rows
pd.set_option('display.max_rows', len(column_info))
print(column_info)
pd.reset_option('display.max_rows')  # Reset option after printing

# Select columns with the highest null values (top 15 for example)
top_null_columns = null_values.nlargest(15).index
top_non_null_values = non_null_values[top_null_columns]
top_null_values = null_values[top_null_columns]

# Plotting
fig, ax = plt.subplots(figsize=(12, 20))
x = df2.columns
ax.barh(x, non_null_values, color="skyblue", label="Non-null values")
ax.barh(x, null_values, left=non_null_values, color="salmon", label="Null values")
ax.set_xlabel("Count")
ax.set_ylabel("Columns")
ax.set_title("Total and Null Values in Each Column in db2")
ax.legend()

plt.tight_layout()
plt.show()
pd.reset_option('display.max_rows')
```
